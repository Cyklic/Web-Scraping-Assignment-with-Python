# **Ethics and Best Practices**
##### Web scraping is a valuable method for gathering data, but it must be approached ethically to protect the interests of website owners and users.
##### The robots.txt file serves as a guideline for websites to indicate which sections are accessible, and which are restricted to web crawlers or scrapers. Ethical scrapers should review and abide by these rules, even if not enforced programmatically, as ignoring them could breach terms of service or cause disruptions.
##### Excessive requests in a short time can overload a website’s server, potentially leading to slow performance or outages for other users. To avoid this, scrapers should control the frequency of their requests, spacing them out to simulate typical user behavior. Rate limiting and adding delays between requests can help maintain server stability, minimize detection, and prevent IP blockage.
##### When scraping content created by users, such as comments or reviews, it is essential to respect privacy. Scrapers should avoid collecting sensitive or personally identifiable information and ensure compliance with regulations like General Data Protection Regulation (GDPR). Any data obtained should be anonymized and used responsibly to avoid harming individuals.
##### In essence, ethical scraping prioritizes respect for the rights of websites and users, ensuring that data collection is conducted in a fair and responsible manner. Adopting these practices helps to maintain trust and reduces potential legal or technical risks.

# **Main Code for Gathering Data**

```python
headset_name = [] #List to store all the headset names
headset_price = [] #List to store all the headset prices
headset_rating = [] #List to store all the headset ratings
headset_review = [] #List to store all the headset review counts

for page in range(1,4): #For loop for pagination, lopping through the page range
    sub_page = 'https://listado.mercadolibre.com.mx/computacion/accesorios-pc-gaming/audifonos/headsets' #Splitting the webpage
    start_index = 1 + (page - 1) * 50 #Handling index for the different pages
    full_page = f'{sub_page}_Desde_{start_index}_NoIndex_True' #Handling new webpage url with index to show pages based on the url

    res = requests.get(full_page) #Getting request

    soup = bs4.BeautifulSoup(res.content, 'html.parser') #Creating soup object

    findings = soup.find_all('div', {'class':'poly-card poly-card--list'}) #Finding all the card for the headsets having the name price ratings discounts etc.



    for result in findings: #For loop for appending all the details found into their corresponding lists
        try:
            headset_name.append(result.find('h2', {'class':'poly-box poly-component__title'}).get_text()) #If found append to headset_name list
        except:
            headset_name.append('N/A') #If not found append 'N/A' to headset_name list
    
        try:
            headset_price.append(result.find('span', {'class':'andes-money-amount andes-money-amount--cents-superscript'}).get_text()) #If found append to headset_price list
        except:
            headset_price.append('NULL') #If not found append 'NULL' to headset_price list
    
        try:
            headset_rating.append(result.find('span', {'class':'poly-reviews__rating'}).get_text()) #If found append to headset_rating list
        except:
            headset_rating.append('0') #If not found append '0' to headset_rating list
    
        try:
            headset_review.append(result.find('span', {'class':'poly-reviews__total'}).get_text()) #If found append to headset_review list
        except:
            headset_review.append('0') #If not found append '0' to headset_review list
```

```python
#Test to see if above loop works
headset_name
```

# **Storing and Cleaning the Data**

```python
#Creating dictionary for storing data to dataframe
headsets = {'Name':headset_name, 'Price$':headset_price, 'AvgRatingoutof5':headset_rating, 'ReviewCount':headset_review}
```

```python
#Storing to dataframe
df = pd.DataFrame(headsets)
df
```

```python
df['Price$'] = df['Price$'].str.replace('$', '', regex=False).str.replace(',', '.', regex=False).astype(float) #Removing dollar sign, changing comma to decimal point, and changing datatype to float from price records
df['AvgRatingoutof5'] = df['AvgRatingoutof5'].astype(float) #Changing ratings datatype to float
df['ReviewCount'] = df['ReviewCount'].str.replace('(', '', regex=False).str.replace(')', '', regex=False).astype(int) #Removing parentheses and changing data type to int
df
```

```python
#To see if there are empty cells
df.isnull().sum()
```

```python
#To see if there are duplicated records
df.duplicated().sum()
```

```python
#Dropping duplicated records
df = df.drop_duplicates()
df
```

```python
#Cross checking duplicates
df.duplicated().sum()
```
